            <strong>1. Remove Rows:</strong><br>

```python

# Fill 'director' and 'cast' with 'Unknown'
df['director'].fillna(df['director'].mode()[0], inplace=True)
df['cast'].fillna(df['cast'].mode()[0], inplace=True)
df['date_added'].fillna('Unknown', inplace=True)


# Fill 'country' with the mode of 'country' for each 'director'
df['country'] = df.groupby('director')['country'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else df['country'].mode()[0]))

# # Backward fill for 'date_added' and 'rating'
# df['date_added'].fillna(method='bfill', inplace=True)
# df['rating'].fillna(method='bfill', inplace=True)


```

```python
df.isna().sum()
```

```python
# Drop rows with any missing values
df.dropna(inplace=True)

# Drop columns with more than a certain threshold of missing values
threshold = 0.5  # e.g., drop columns with more than 50% missing values
df.dropna(thresh=int(df.shape[0] * threshold), axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df['country'] = df['country'].apply(lambda x:x.capitalize())
df['country'].head()
```

```python
data_sub = df.groupby('type')['month'].value_counts().unstack().fillna(0).loc[['TV Show', 'Movie']].cumsum(axis=0).T

data_sub2 = data_sub
data_sub2['Value'] = data_sub2['Movie'] + data_sub2['TV Show']
data_sub2 = data_sub2.reset_index()

data_sub2['month'] = pd.Categorical(data_sub2['month'], categories=[
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'], ordered=True)
df_polar = data_sub2.sort_values(by='month', ascending=True)

color_map = ['#221f1f' for _ in range(12)]
color_map[0] = color_map[11] = '#b20710'  # ŸÑŸàŸÜ ÿßŸÑÿ™ŸÖŸäŸäÿ≤

 
max_value = df_polar['Value'].max()
slope = (30 - 1) / max_value
heights = slope * df_polar['Value'] + 1

width = 2 * np.pi / len(df_polar.index)
angles = [element * width for element in range(len(df_polar.index))] 
fig = go.Figure()

fig.add_trace(go.Barpolar(
    r=heights,
    theta=[np.degrees(angle) for angle in angles],
    width=[np.degrees(width)] * len(df_polar.index),
    marker=dict(
        color=color_map,
        line=dict(
            color='white',
            width=2
        )
    ),
    opacity=0.8
))

# ÿ•ÿ∂ÿßŸÅÿ© ÿßŸÑŸÜÿµŸàÿµ
for angle, height, label in zip(angles, heights, df_polar["month"]):
    rotation = np.degrees(angle)
    alignment = "right" if np.pi / 2 <= angle < 3 * np.pi / 2 else "left"
    if alignment == "right":
        rotation += 180

    fig.add_annotation(
        x=np.degrees(angle),
        y=height + 2,
        text=label,
        showarrow=False,
        font=dict(size=12, family='Arial', color='black'),
        textangle=rotation if alignment == "left" else rotation - 180,
        xanchor=alignment,
        yanchor='middle'
    )

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=False),
        angularaxis=dict(
            tickvals=[0, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330],
            ticktext=['January', 'February', 'March', 'April', 'May', 'June',
                      'July', 'August', 'September', 'October', 'November', 'December'],
            tickmode='array',
            rotation=0,
            direction='clockwise'
        )
    ),
    showlegend=False,
    title="Movies & TV Shows Added Over Time",
    title_font_size=20,
    width=800,
    height=800
)

fig.show()
```
